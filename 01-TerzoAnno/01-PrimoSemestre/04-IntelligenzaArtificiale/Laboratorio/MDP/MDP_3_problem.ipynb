## Assignment 1: Value Iteration Algorithm

Your first assignment is to implement the Value Iteration algorithm on LavaFloor. The solution returned by your algorithm must be a 1-d array of action identifiers where the $i$-th action refers to the $i$-th state.  You can perform all the test on a different versions of the environment, but with the same structure: *HugeLavaFloor*, *NiceLavaFloor* and *VeryBadLavaFloor*.

<img src="images/value-iteration.png" width="600">

The *value_iteration* function has to be implemented. Notice that the value iteration approach return a matrix with the value for eacht state, the function *values_to_policy* automatically convert this matrix in the policy.

```python
def value_iteration(environment, maxiters=300, discount=0.9, max_error=1e-3):
    """
    Performs the value iteration algorithm for a specific environment
    
    Args:
        environment: OpenAI Gym environment
        maxiters: timeout for the iterations
        discount: gamma value, the discount factor for the Bellman equation
        max_error: the maximum error allowd in the utility of any state
        
    Returns:
        policy: 1-d dimensional array of action identifiers where index `i` corresponds to state id `i`
        utilities: 1-d dimensional array of action utilities
    """

    #Initialize the utility vector for all states
    num_states = environment.observation_space.n
    num_actions = environment.action_space.n
    U_1 = [0.0 for _ in range(num_states)]
    
    #Calculate convergence threshold based on the Bellman error
    threshold = max_error * ((1 - discount) / discount)
    
    for i in range(maxiters):
        #Create a copy of the current utilities for reference
        U = U_1.copy()
        delta = 0
        
        for state in range(num_states):
            #Check for terminal states (Pit or Goal) in the grid
            if environment.grid[state] in ["P", "G"]:
                # Terminal states take the immediate reward value directly
                U_1[state] = environment.RS[state]
                
            else:
                #Calculate expected utility for non-terminal states
                action_utilities = []
                
                for action in range(num_actions):
                    u_sum = 0
                    # Iterate over all potential next states based on Transition Matrix T
                    for next_state in range(num_states):
                        prob = environment.T[state, action, next_state]
                        # Accumulate expected utility: P(s'|s,a) * U(s')
                        u_sum += prob * U[next_state]
                        
                    action_utilities.append(u_sum)
                
                #Apply Bellman Equation: R(s) + discount * max(expected_utilities)
                U_1[state] = environment.RS[state] + discount * max(action_utilities)
            
            #Update the maximum variation (delta) seen in this iteration
            current_diff = abs(U_1[state] - U[state])
            if current_diff > delta: delta = current_diff
        
        #Check for convergence
        if delta < threshold:
            break

    #Return derived policy and the final utility vector
    return values_to_policy(np.asarray(U_1), environment), U_1
```

## Assignment 2: Policy Iteration Algorithm (<span style="color:red">*OPTIONAL*</span>)

**The following function has to be implemented:**

```python
def argmax(a):
    return max(range(len(a)), key=lambda x : a[x])
```

```python
def policy_iteration(environment, maxiters=150, discount=0.9, maxviters=10, max_error=1e-3):
    """
    Performs the policy iteration algorithm for a specific environment
    
    Args:
        environment: OpenAI Gym environment
        maxiters: timeout for the iterations
        discount: gamma value, the discount factor for the Bellman equation
        
    Returns:
        policy: 1-d dimensional array of action identifiers where index `i` corresponds to state id `i`
    """

    # Initialize policy (arbitrary action 0) and utility vector
    num_states = environment.observation_space.n
    num_actions = environment.action_space.n
    
    policy = [0 for _ in range(num_states)] 
    U = [0.0 for _ in range(num_states)]
    
    iter_count = 0
    
    # Main loop: Policy Improvement
    while True:
        # --- Step 1: Policy Evaluation ---
        # Evaluate the utility of the current policy until convergence or maxviters
        for v in range(maxviters):
            delta = 0
            U_prev = U.copy()
            
            for state in range(num_states):
                # Handle Terminal States (Pit/Goal)
                if environment.grid[state] in ["P", "G"]:
                    U[state] = environment.RS[state]
                    continue
                
                # Calculate utility based on the specific action chosen by the current policy
                chosen_action = policy[state]
                u_sum = 0
                for next_state in range(num_states):
                    # Bellman Expectation Equation: Sum(T * U_prev)
                    u_sum += environment.T[state, chosen_action, next_state] * U_prev[next_state]
                
                U[state] = environment.RS[state] + discount * u_sum
                
                # Track max change for convergence
                diff = abs(U[state] - U_prev[state])
                if diff > delta:
                    delta = diff
            
            # Stop evaluation if utilities define the policy well enough
            if delta < max_error * (1 - discount) / discount:
                break

        # --- Step 2: Policy Improvement ---
        policy_stable = True
        
        for state in range(num_states):
            # Skip improvement for terminal states
            if environment.grid[state] in ["P", "G"]:
                continue
            
            old_action = policy[state]
            action_utilities = []
            
            # Calculate Q-value for every possible action in this state
            for action in range(num_actions):
                val = 0
                for next_state in range(num_states):
                    val += environment.T[state, action, next_state] * U[next_state]
                action_utilities.append(val)
            
            # Greedily choose the best action
            best_action = np.argmax(action_utilities)
            
            # Update policy if a better action is found
            if best_action != old_action:
                policy[state] = best_action
                policy_stable = False
        
        iter_count += 1
        
        # Terminate if policy hasn't changed or max iterations reached
        if policy_stable or iter_count >= maxiters:
            break
    
    # Return matched to the previous format: (policy, utilities)
    return np.asarray(policy), np.asarray(U)
```
